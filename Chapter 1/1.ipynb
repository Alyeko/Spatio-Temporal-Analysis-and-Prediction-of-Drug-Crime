```python
import pandas as pd
import geopandas as gpd
import contextily as cx
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar
from matplotlib.font_manager import FontProperties
```

```python
westminster_wards_w_crime = gpd.read_file('westminster.shp')
westminster_wards_w_crime
```

```python
westmstr = westminster_wards_w_crime.groupby('Ward Name').sum()
```

```python
westmstr
westmstr.drop(['HECTARES'], axis=1, inplace=True)
```

```python
westmstr
```

```python
westmstr['Drug Crime Total']= westmstr.iloc[:, 0:].sum(axis=1)
crime_total = westmstr['Drug Crime Total']
crime_total
```

```python
crime_total_df = crime_total.to_frame().reset_index()
crime_total_df.sort_values(by='Drug Crime Total', ascending=False)
```

```python
ldn_wards = gpd.read_file('London-wards-2018_ESRI/London_Ward.shp')
city_of_westminster = ldn_wards[ldn_wards['DISTRICT'] == 'City of Westminster']
city_of_westminster = city_of_westminster.rename(columns = {'NAME':'Ward Name'})
city_of_westminster = city_of_westminster.merge(crime_total_df, on='Ward Name')
```

```python
city_of_westminster.columns
```

```python
plt.rcParams['figure.figsize'] = [15,10]
```

```python
west_city_label_df = city_of_westminster.sort_values(by='Ward Name',  key=lambda x: x.str.len()) #########
west_city_label_df = west_city_label_df.reset_index()
west_city_label_df
```

```python
ax = city_of_westminster.plot(column = 'Drug Crime Total', scheme='natural_breaks', cmap = 'Greys', edgecolor = 'purple', legend=True,
              legend_kwds={'loc': 'upper right', 'fmt':'{:.0f}', 'fontsize': 13, 
                              'title': 'DRUG CRIME COUNT', 'frameon': True, 'facecolor' : 'white', 
                              'edgecolor': 'purple',
                              'labels' : ['[145 - 464]', '[464 - 1038]', '[1038 - 2077]', '[2077 - 6148]' ,'[6148 - 7336]']})

x, y, arrow_length = 0.035, 0.96, 0.1
ax.annotate('N', xy=(x, y), xytext=(x, y-arrow_length),
            arrowprops=dict(facecolor='black', width=15, headwidth=35),
            ha='center', va='center', fontsize=20,
            xycoords=ax.transAxes) 

west_city_label_df[:14].apply(lambda x: ax.annotate(text=x['Ward Name'], xy=x.geometry.centroid.coords[0], ha='center', fontsize=10), axis=1)
west_city_label_df[14:].apply(lambda x: ax.annotate(text=x['index'], xy=x.geometry.centroid.coords[0], ha='center', fontsize=10), axis=1)
west_city_label_df.iloc[[9]].apply(lambda x: ax.annotate(text=x['Ward Name'], xy=x.geometry.centroid.coords[0], ha='center', color='white', fontsize=10), axis=1)

colLabels = ['Label', 'Ward Name']
label_name = west_city_label_df[14:][['index', 'Ward Name']].sort_values(by='index')
tab = ax.table(cellText=label_name.values, colLabels=colLabels, colWidths = [0.1,0.2], loc='lower left', 
               colLoc = 'right',)
for (row, col), cell in tab.get_celld().items():
     cell.set_text_props(fontproperties=FontProperties(weight='normal'),)
tab.scale(1.6, 1.6)

scalebar = ScaleBar(0.2, location='lower right') # 1 pixel = 0.2 meter
plt.gca().add_artist(scalebar)
plt.title("A map of the borough of Westminster, London showing the accumulated drug crime counts in all wards from 2011 to 2018")
plt.rcParams['figure.figsize'] = [15,10]
```
