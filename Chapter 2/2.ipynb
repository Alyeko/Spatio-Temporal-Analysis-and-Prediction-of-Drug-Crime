### 2.3 Time-series plot

```python
from pandas import read_csv
from matplotlib import pyplot
pyplot.rcParams['figure.figsize']=[10, 3]
# load dataset
series = read_csv('st_james.csv', header=0, index_col=0)
values = series.values
# plot dataset
pyplot.plot(values, color = 'purple')
pyplot.rcParams['figure.figsize']=[10, 3]
pyplot.title("St. James's monthly drug crime count from 2011 to 2018")
pyplot.xlabel('Time (months)')
pyplot.ylabel('Drug Crime Count')
pyplot.show()
```

```python
series_new = series.reset_index()
series_new
```

### 2.4 Time-series decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result=seasonal_decompose(values, model='additive', period=12)
```

```python
fig, axes = pyplot.subplots(4, 1, sharex=True)


axes[0].plot(result.observed, color='purple')
axes[0].set_ylabel('Observed')

axes[1].plot(result.trend, color='purple')
axes[1].set_ylabel('Trend')

axes[2].plot(result.seasonal, color='purple')
axes[2].set_ylabel('Seasonal')

axes[3].plot(result.resid,  marker='o', color='purple')
axes[3].set_ylabel('Residual')
pyplot.xlabel('Time (months)')

```
